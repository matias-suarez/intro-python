### Listas, tuplas y diccionarios

Las listas y tuplas son secuencias de numeros o strings. **LA DIFERENCIA** es que las **listas son mutables**. Es decir, es un objeto que puede cambiar: extenderse con otra secuencia, agregar o quitar elementos, cambiar un elemento o una porción por otra, reordenarse *in place*, etc.  

Los diccionarios son otro tipo de estructuras de alto nivel que ya vienen incorporados. A diferencia de las secuencias, los valores **no están en una posición** sino bajo **una clave**: son asociaciones `clave:valor`

### **Listas**

````python
# Ejemplo de lista

lista_ejemplo = [1,2,-1,0,'hola']

print(type(lista_ejemplo))

# Podemos ver los elementos que hay en cada posición
print(lista_ejemplo[0]) # 0 es la primera posición
print(lista_ejemplo[-1]) # -1 es la ultima posición
````

### **Tuplas**

````python
# Ejemplo de tupla

tupla_ejemplo = (1,2,-1,0,'hola')

print(type(tupla_ejemplo))

# Podemos ver los elementos que hay en cada posición
print(tupla_ejemplo[0]) # 0 es la primera posición
print(tupla_ejemplo[-1]) # -1 es la ultima posición
````

````python
# Es posible modificar el valor de un elemento de la lista
# Por ejemplo cambio el valor del primer elemento por 10
lista_ejemplo[0] = 10

print(lista_ejemplo)
````

````python
# NO es posible modificar el valor de un elemento de una tupla
# Por ejemplo intento cambiar el valor del primer elemento por 10
# tupla_ejemplo[0] = 10

print(tupla_ejemplo)
````

Los diccionarios **son mutables**. Es decir, podemos cambiar el valor de una clave, agregar o quitar.

### Estructuras Lógicas IF

Son estructuras que, dada una o varias condiciones, permiten ejecutar secciones de códigos.

Operadores:

#### == igual a

#### != distinto de

#### < menor que

#### > mayor que

#### <= menor igual a

#### >= mayor igual a

````python
# Por ejemplo: Queremos que dado un valor de x, si este es menor a 5 imprime en
# pantalla que el valor es menor a 5, de lo contrario imprime que es mayor a 5

x = 3

if x <= 5: # Si x es menor igual a 5 entra en esta parte del codigo
  print('El valor de x es menor igual a 5')
else: # de lo contrario entra en esta parte del codigo
  print('El valor de x es mayor a 5')
````

### Podemos agregar mas de una condicion:

````python
x = 50

if x <= 0: # Si x es menor igual a 0 entra en esta parte del codigo
  print('El valor de x es menor igual a 0')
elif 0 < x <= 10: # Si x es mayor que 0 pero menor igual a 10 entra en esta parte del codigo
  print('El valor de x es mayor que 0 pero menor igual a 10')
elif 10 < x <= 100: # Si x es mayor que 0 pero menor igual a 10 entra en esta parte del codigo
  print('El valor de x es mayor que 10 pero menor igual a 100')
else: # de lo contrario entra en esta parte del codigo
  print('El valor de x es mayor a 100')
````

### Podemos combinar lógica booleana con estructura IF

````python
condicion = True

if condicion == True:
  print('La condicion es verdadera')
else:
  print('La condicion es falsa')
````

````python
# Equivalentemente podemos definir el bloque IF anterior de esta manera:
if condicion:
  print('La condicion es verdadera')
else:
  print('La condicion es falsa')
````

### Funciones

*Las funciones se definen con la palabra clave def*

**Sintaxis**

def nombre_funcion(parametros):
    
    Cuerpo de la función
    
    return

````python
# Definimos una funcion que suma dos valores y regresa el resultado

def suma(a,b):
  c = a + b
  return c
````

````python
# Llamamos a la función suma

suma(3,4)
````

````python
# Podemos definir una función con valores por defecto. Es decir, valores
# que si no son especificados son tomados por defecto con un valor asignado

def suma_triple(a,b,c=0):
  d = a + b + c
  return d
````

````python
# Llamamos a la función suma

suma_triple(3,4) # Si no especificamos c se lo considera como c=3
````

````python
suma_triple(3,4,2) # Si lo especificamos lo considera con el valor que le damos de entrada
````

````python
lista1 = [1,2,3,4,5]
lista2 = [-1,2,-3,4,0]

# Podemos darle a una funcion otros objetos como argumentos:
# Llamamos a la función

suma(lista1,lista2)

# Como resultado obtenemos una lista concatenada
````

### Operadores Lógicos AND, OR, IN, NOT IN


````python
# Si queremos ejecutar un bloque de código si se cumplen dos o mas condiciones
# usamos AND, si solo queremos ejecutar si se cumple una sola condición (y las otras no)
# usamos OR

x = 5; y = 20

if x == 5 and y == 20:
  print('x es igual a 5 e y es igual a 20')
else:
  print('x es distinto de 5 o y es distinto de 20')
````

````python
x = 3; y = 10

if x == 5 or y == 20:
  print('x es igual a 5 o y es igual a 20')
else:
  print('x es distinto de 5 e y es distinto de 20 (simultaneamente)')
````

````python
# Si queremos verificar, por ejemplo, si un elemento está o no en
# una lista usamos IN y NOT IN

x_list = [0,1,2,3,4,5,6]

print(1 in x_list, -1 in x_list)
````

````python
# Analogamente

print(1 not in x_list, -1 not in x_list)
````

## Pandas


La librería pandas en Python es una herramienta de manipulación y análisis de datos de alto rendimiento. Ofrece estructuras de datos flexibles y eficientes, diseñadas para facilitar la manipulación y análisis de datos en Python. Las dos estructuras de datos principales en pandas son Series y DataFrame.

*  Series: Una Serie es un objeto unidimensional similar a un array que puede
contener cualquier tipo de datos, como números enteros, flotantes, cadenas, etc. Cada elemento en una Serie tiene una etiqueta asociada, denominada índice.

*  DataFrame: Un DataFrame es una estructura de datos bidimensional similar a una tabla, que consta de filas y columnas. Puedes pensar en un DataFrame como una hoja de cálculo o una tabla SQL. Cada columna en un DataFrame es una Serie.

Además de estas estructuras de datos, pandas proporciona una amplia gama de herramientas para realizar operaciones de limpieza, transformación y análisis de datos, como filtrado, agrupamiento, pivotado, fusión, concatenación, entre otras.

Pandas es una de las herramientas más populares en el ámbito de la ciencia de datos en Python debido a su facilidad de uso y su potencia para manejar datos de forma eficiente. Es frecuentemente utilizada en tareas como manipulación de datos, limpieza de datos, análisis exploratorio, y preparación de datos para su modelado y visualización.

````python
# Primero importamos la librería pandas
import pandas as pd
# En caso de no estar instalada, se la puede instalar con pip install pandas
````

### DateTime de Pandas

El módulo DateTime nos permite trabajar con series temporales en Pandas. Las principales funciones son:

1. **Conversión de Fechas**: Permite convertir strings y otros formatos en objetos de fecha y hora.

2. **Generación de Rangos de Fechas**: Facilita la creación de secuencias de fechas con especificación de inicio, fin y frecuencia.

3. **Representación de Tiempos Específicos**: Ofrece la posibilidad de trabajar con momentos concretos en el tiempo.

4. **Duraciones**: Permite representar y operar con diferencias entre fechas, lo que resulta útil para cálculos de tiempo.

5. **Acceso a Componentes de Fechas**: Proporciona acceso a atributos como año, mes, día y hora en series de fechas.

6. **Desplazamiento de Fechas**: Facilita la suma de períodos de tiempo específicos, como meses o años, a objetos de fecha.

7. **Formateo de Fechas**: Permite convertir fechas y horas a cadenas en formatos personalizados, o viceversa.

8. **Re-muestreo de Series Temporales**: Permite cambiar la frecuencia de los datos temporales, como transformar datos diarios en mensuales.

pd.to_datetime(): Convierte eficientemente una representación de cadena de una fecha en un formato de fecha real, lo cual es útil al aprovechar la vasta funcionalidad de fechas proporcionada por Pandas, como el remuestreo.

````python
pd.to_datetime(['2021-01-01', '2021-02-01'])
````

pd.date_range(): Genera un rango de fechas. Permite especificar el inicio, el final y la frecuencia (diaria, mensual, etc.).

````python
pd.date_range(start='2021-01-01', end='2021-01-10', freq='D')
````

pd.Timestamp: Representa un punto específico en el tiempo y se puede usar para crear objetos de fecha y hora.

````python
ts = pd.Timestamp('2021-01-01')
ts
````

pd.Timedelta: Representa una duración, la diferencia entre dos fechas. Es útil para sumar o restar tiempo a un objeto Timestamp.

````python
td = pd.Timedelta(days=1)
td
````

Ejemplo de uso de Timedelta

````python
ts + td
````

Acceso a atributos de fecha: Una vez que tienes una serie de fechas, puedes acceder a atributos como el año, mes, día, hora, minuto, etc.

## Matplotlib

La librería Matplotlib es una biblioteca de Python que proporciona una amplia gama de herramientas para crear visualizaciones estáticas, interactivas y animadas. Con Matplotlib, puedes generar una variedad de gráficos, como gráficos de líneas, gráficos de dispersión, histogramas, gráficos de barras, gráficos de contorno, gráficos de superficie y muchos más.

Matplotlib se puede utilizar tanto para crear gráficos simples como para producir visualizaciones complejas y personalizadas. La flexibilidad de Matplotlib permite controlar casi todos los aspectos de la apariencia de los gráficos, incluidos colores, estilos de línea, etiquetas de ejes, títulos, leyendas y más.

Además de su funcionalidad básica, Matplotlib se integra bien con otras bibliotecas de Python, como NumPy y Pandas, lo que facilita la visualización de datos almacenados en estructuras de datos comunes. También es posible combinar Matplotlib con herramientas de GUI como Tkinter, PyQt o wxPython para crear aplicaciones con interfaces gráficas interactivas.

En resumen, Matplotlib es una herramienta fundamental en el ecosistema de Python para visualización de datos y trazado de gráficos, utilizada ampliamente en campos como la ciencia de datos, la ingeniería, la investigación académica y muchos otros.

````python
# Importamos la librería
import matplotlib.pyplot as plt
````

````python
# Creamos un vector x y una función y=f(x) para graficar
x = np.linspace(0, 10, 11)
y = x ** 2
````

````python
# Chequeamos los valore de x e y
print(x)
print(y)
````

### Grafico simple

````python
# Graficamos
plt.plot(x,y)
````

````python
# Podemos agregar nombres a los ejes x e y. Incluso podemos agregar título al gráfico
plt.plot(x,y)
plt.xlabel('Nombre eje x')
plt.ylabel('Nombre eje y')
plt.title('Título')
plt.show()
````

````python
# Podemos agregar una leyenda y cambiar el color de la curva
plt.plot(x, y, color='red', label='Curva')
plt.legend()
plt.show()
````

### Múltiples Gráficos


````python
# Tambien podemos hacer dos o más gráficos en una misma ventana

plt.subplot(1,2,1)
plt.plot(x, y, color='red', linestyle='--')

plt.subplot(1,2,2)
plt.plot(y, x, color='green', linestyle='dashdot')

plt.show()
````

````python
# Tambien podemos crear un gráfico de la siguiente manera:
# Creamos Figura
fig = plt.figure()

# Agregamos los ejes a la figura
ax = fig.add_axes([0, 0, 0.8, 0.8]) # izq, abajo, ancho, alto (van de 0 a 1)

# Graficamos en los ejes anteriores
ax.plot(x, y, 'b')
ax.set_xlabel('Eje x') # Notar el uso de set_ para colocar los nombres de los ejes y titulos
ax.set_ylabel('Eje y')
ax.set_title('Titulo')
````

De la forma anterior el código es un poco más complicado pero la ventaja es que ahora tenemos control completo de la ubicación de los ejes y podremos agregar facilmente más de un eje a una misma figura:

````python
# Creamos figura
fig = plt.figure()

ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # eje principal
ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # eje secundario

# Figura grande o principal
ax1.plot(x, y, 'b')
ax1.set_xlabel('Eje x principal')
ax1.set_ylabel('Eje y principal')
ax1.set_title('Titulo principal')

# Fgiura pequeña o secundaria
ax2.plot(y, x, 'r')
ax2.set_xlabel('Eje x secundario')
ax2.set_ylabel('Eje y secundario')
ax2.set_title('Titulo secundario')
````

````python
# Creamos multiples gráficos en una misma figura
fig, ax = plt.subplots(nrows=2, ncols=2)

ax[0,0].plot(x, y, 'g')
ax[0,1].plot(x, y, 'b')
ax[1,0].plot(x, y, 'r')
ax[1,1].plot(x, y, 'y')

# ax[fila, columna]
````

````python
# O tambien
fig, ax = plt.subplots()

ax.plot(x, x**2, 'b.-') # linea azul con puntos
ax.plot(x, x**3, 'g--') # linea verde discontinua
ax.plot(x, x**0.5, 'r-') # linea roja continua
````

### Estilos personalizados para graficar en matplotlib

Distintos tipos de estilos para graficar

````python
fig, ax = plt.subplots(figsize=(12,6))

# Distintos ancho de linea
ax.plot(x, x+1, color="red", linewidth=0.25)
ax.plot(x, x+2, color="red", linewidth=0.50)
ax.plot(x, x+3, color="red", linewidth=1.00)
ax.plot(x, x+4, color="red", linewidth=2.00)

# opciones para estilo de linea ‘-‘, ‘–’, ‘-.’, ‘:’, ‘steps’
ax.plot(x, x+5, color="green", lw=3, linestyle='-')
ax.plot(x, x+6, color="green", lw=3, ls='-.')
ax.plot(x, x+7, color="green", lw=3, ls=':')

# discontinuidad personalizada
line, = ax.plot(x, x+8, color="black", lw=1.50)
line.set_dashes([5, 10, 15, 10]) # formato: longitud linea, longitud espaciado, ...

# posibles opciones para marcadores: marcadores = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x, x+ 9, color="blue", lw=3, ls='-', marker='+')
ax.plot(x, x+10, color="blue", lw=3, ls='--', marker='o')
ax.plot(x, x+11, color="blue", lw=3, ls='-', marker='s')
ax.plot(x, x+12, color="blue", lw=3, ls='--', marker='1')

# tamaño y color del marcador
ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
ax.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
ax.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8,
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green")
````
