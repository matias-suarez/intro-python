### Estructuras Lgicas IF

Son estructuras que, dada una o varias condiciones, permiten ejecutar secciones de cdigos.

Operadores:

#### == igual a

#### != distinto de

#### < menor que

#### > mayor que

#### <= menor igual a

#### >= mayor igual a

````python
# Por ejemplo: Queremos que dado un valor de x, si este es menor a 5 imprime en
# pantalla que el valor es menor a 5, de lo contrario imprime que es mayor a 5

x = 3

if x <= 5: # Si x es menor igual a 5 entra en esta parte del codigo
  print('El valor de x es menor igual a 5')
else: # de lo contrario entra en esta parte del codigo
  print('El valor de x es mayor a 5')
````

### Podemos agregar mas de una condicion:

````python
x = 50

if x <= 0: # Si x es menor igual a 0 entra en esta parte del codigo
  print('El valor de x es menor igual a 0')
elif 0 < x <= 10: # Si x es mayor que 0 pero menor igual a 10 entra en esta parte del codigo
  print('El valor de x es mayor que 0 pero menor igual a 10')
elif 10 < x <= 100: # Si x es mayor que 0 pero menor igual a 10 entra en esta parte del codigo
  print('El valor de x es mayor que 10 pero menor igual a 100')
else: # de lo contrario entra en esta parte del codigo
  print('El valor de x es mayor a 100')
````

### Podemos combinar lgica booleana con estructura IF

````python
condicion = True

if condicion == True:
  print('La condicion es verdadera')
else:
  print('La condicion es falsa')
````

````python
# Equivalentemente podemos definir el bloque IF anterior de esta manera:
if condicion:
  print('La condicion es verdadera')
else:
  print('La condicion es falsa')
````

### Funciones

*Las funciones se definen con la palabra clave def*

**Sintaxis**

````python
# Definimos una funcion que suma dos valores y regresa el resultado

def suma(a,b):
  c = a + b
  return c
````

````python
# Llamamos a la funcin suma

suma(3,4)
````

````python
# Podemos definir una funcin con valores por defecto. Es decir, valores
# que si no son especificados son tomados por defecto con un valor asignado

def suma_triple(a,b,c=0):
  d = a + b + c
  return d
````

````python
# Llamamos a la funcin suma

suma_triple(3,4) # Si no especificamos c se lo considera como c=3
````

````python
suma_triple(3,4,2) # Si lo especificamos lo considera con el valor que le damos de entrada
````

````python
lista1 = [1,2,3,4,5]
lista2 = [-1,2,-3,4,0]

# Podemos darle a una funcion otros objetos como argumentos:
# Llamamos a la funcin

suma(lista1,lista2)

# Como resultado obtenemos una lista concatenada
````

### Operadores Lgicos AND, OR, IN, NOT IN


````python
# Si queremos ejecutar un bloque de cdigo si se cumplen dos o mas condiciones
# usamos AND, si solo queremos ejecutar si se cumple una sola condicin (y las otras no)
# usamos OR

x = 5; y = 20

if x == 5 and y == 20:
  print('x es igual a 5 e y es igual a 20')
else:
  print('x es distinto de 5 o y es distinto de 20')
````

````python
x = 3; y = 10

if x == 5 or y == 20:
  print('x es igual a 5 o y es igual a 20')
else:
  print('x es distinto de 5 e y es distinto de 20 (simultaneamente)')
````

````python
# Si queremos verificar, por ejemplo, si un elemento est o no en
# una lista usamos IN y NOT IN

x_list = [0,1,2,3,4,5,6]

print(1 in x_list, -1 in x_list)
````

````python
# Analogamente

print(1 not in x_list, -1 not in x_list)
````

## Pandas


La librera pandas en Python es una herramienta de manipulacin y anlisis de datos de alto rendimiento. Ofrece estructuras de datos flexibles y eficientes, diseadas para facilitar la manipulacin y anlisis de datos en Python. Las dos estructuras de datos principales en pandas son Series y DataFrame.

*  Series: Una Serie es un objeto unidimensional similar a un array que puede
contener cualquier tipo de datos, como nmeros enteros, flotantes, cadenas, etc. Cada elemento en una Serie tiene una etiqueta asociada, denominada ndice.

*  DataFrame: Un DataFrame es una estructura de datos bidimensional similar a una tabla, que consta de filas y columnas. Puedes pensar en un DataFrame como una hoja de clculo o una tabla SQL. Cada columna en un DataFrame es una Serie.

Adems de estas estructuras de datos, pandas proporciona una amplia gama de herramientas para realizar operaciones de limpieza, transformacin y anlisis de datos, como filtrado, agrupamiento, pivotado, fusin, concatenacin, entre otras.

Pandas es una de las herramientas ms populares en el mbito de la ciencia de datos en Python debido a su facilidad de uso y su potencia para manejar datos de forma eficiente. Es frecuentemente utilizada en tareas como manipulacin de datos, limpieza de datos, anlisis exploratorio, y preparacin de datos para su modelado y visualizacin.

### DateTime de Pandas

El mdulo DateTime nos permite trabajar con series temporales en Pandas. Las principales funciones son:

1. **Conversin de Fechas**: Permite convertir strings y otros formatos en objetos de fecha y hora.

2. **Generacin de Rangos de Fechas**: Facilita la creacin de secuencias de fechas con especificacin de inicio, fin y frecuencia.

3. **Representacin de Tiempos Especficos**: Ofrece la posibilidad de trabajar con momentos concretos en el tiempo.

4. **Duraciones**: Permite representar y operar con diferencias entre fechas, lo que resulta til para clculos de tiempo.

5. **Acceso a Componentes de Fechas**: Proporciona acceso a atributos como ao, mes, da y hora en series de fechas.

6. **Desplazamiento de Fechas**: Facilita la suma de perodos de tiempo especficos, como meses o aos, a objetos de fecha.

7. **Formateo de Fechas**: Permite convertir fechas y horas a cadenas en formatos personalizados, o viceversa.

8. **Re-muestreo de Series Temporales**: Permite cambiar la frecuencia de los datos temporales, como transformar datos diarios en mensuales.

pd.to_datetime(): Convierte eficientemente una representacin de cadena de una fecha en un formato de fecha real, lo cual es til al aprovechar la vasta funcionalidad de fechas proporcionada por Pandas, como el remuestreo.

````python
pd.to_datetime(['2021-01-01', '2021-02-01'])
````

pd.date_range(): Genera un rango de fechas. Permite especificar el inicio, el final y la frecuencia (diaria, mensual, etc.).

````python
pd.date_range(start='2021-01-01', end='2021-01-10', freq='D')
````

pd.Timestamp: Representa un punto especfico en el tiempo y se puede usar para crear objetos de fecha y hora.

````python
ts = pd.Timestamp('2021-01-01')
ts
````

pd.Timedelta: Representa una duracin, la diferencia entre dos fechas. Es til para sumar o restar tiempo a un objeto Timestamp.

````python
td = pd.Timedelta(days=1)
td
````

## Matplotlib

La librera Matplotlib es una biblioteca de Python que proporciona una amplia gama de herramientas para crear visualizaciones estticas, interactivas y animadas. Con Matplotlib, puedes generar una variedad de grficos, como grficos de lneas, grficos de dispersin, histogramas, grficos de barras, grficos de contorno, grficos de superficie y muchos ms.

Matplotlib se puede utilizar tanto para crear grficos simples como para producir visualizaciones complejas y personalizadas. La flexibilidad de Matplotlib permite controlar casi todos los aspectos de la apariencia de los grficos, incluidos colores, estilos de lnea, etiquetas de ejes, ttulos, leyendas y ms.

Adems de su funcionalidad bsica, Matplotlib se integra bien con otras bibliotecas de Python, como NumPy y Pandas, lo que facilita la visualizacin de datos almacenados en estructuras de datos comunes. Tambin es posible combinar Matplotlib con herramientas de GUI como Tkinter, PyQt o wxPython para crear aplicaciones con interfaces grficas interactivas.

En resumen, Matplotlib es una herramienta fundamental en el ecosistema de Python para visualizacin de datos y trazado de grficos, utilizada ampliamente en campos como la ciencia de datos, la ingeniera, la investigacin acadmica y muchos otros.

### Mltiples Grficos


````python
# Tambien podemos hacer dos o ms grficos en una misma ventana

plt.subplot(1,2,1)
plt.plot(x, y, color='red', linestyle='--')

plt.subplot(1,2,2)
plt.plot(y, x, color='green', linestyle='dashdot')

plt.show()
````

````python
# Tambien podemos crear un grfico de la siguiente manera:
# Creamos Figura
fig = plt.figure()

# Agregamos los ejes a la figura
ax = fig.add_axes([0, 0, 0.8, 0.8]) # izq, abajo, ancho, alto (van de 0 a 1)

# Graficamos en los ejes anteriores
ax.plot(x, y, 'b')
ax.set_xlabel('Eje x') # Notar el uso de set_ para colocar los nombres de los ejes y titulos
ax.set_ylabel('Eje y')
ax.set_title('Titulo')
````

De la forma anterior el cdigo es un poco ms complicado pero la ventaja es que ahora tenemos control completo de la ubicacin de los ejes y podremos agregar facilmente ms de un eje a una misma figura:

````python
# Creamos figura
fig = plt.figure()

ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # eje principal
ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # eje secundario

# Figura grande o principal
ax1.plot(x, y, 'b')
ax1.set_xlabel('Eje x principal')
ax1.set_ylabel('Eje y principal')
ax1.set_title('Titulo principal')

# Fgiura pequea o secundaria
ax2.plot(y, x, 'r')
ax2.set_xlabel('Eje x secundario')
ax2.set_ylabel('Eje y secundario')
ax2.set_title('Titulo secundario')
````

````python
# Creamos multiples grficos en una misma figura
fig, ax = plt.subplots(nrows=2, ncols=2)

ax[0,0].plot(x, y, 'g')
ax[0,1].plot(x, y, 'b')
ax[1,0].plot(x, y, 'r')
ax[1,1].plot(x, y, 'y')

# ax[fila, columna]
````

````python
# O tambien
fig, ax = plt.subplots()

ax.plot(x, x**2, 'b.-') # linea azul con puntos
ax.plot(x, x**3, 'g--') # linea verde discontinua
ax.plot(x, x**0.5, 'r-') # linea roja continua
````
